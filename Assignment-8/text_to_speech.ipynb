```python
# Cell 2: Import and Setup
from TTS.api import TTS
import IPython.display as ipd

tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC")

print("✓ TTS model loaded")
```

```python
# Cell 3: Generate Speech
text = "Hello! This is an example of text to speech conversion using Tacotron."
tts.tts_to_file(text=text, file_path="output.wav")

print("✓ Speech generated and saved as 'output.wav'")
```

```python
# Cell 4: Play Audio
ipd.Audio("output.wav")
```

```python
# Cell 5: Batch TTS
texts = [
    "Artificial intelligence is transforming our world.",
    "Machine learning enables computers to learn from data.",
    "Natural language processing helps computers understand human language.",
]

for idx, text in enumerate(texts):
    filename = f"speech_{idx}.wav"
    tts.tts_to_file(text=text, file_path=filename)
    print(f"✓ Generated: {filename}")

print("\n✓ All speech files generated!")
```
