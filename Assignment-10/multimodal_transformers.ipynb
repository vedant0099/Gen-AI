```python
# Cell 2: Import Libraries
from transformers import CLIPModel, CLIPProcessor
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
import matplotlib.pyplot as plt

print("✓ Libraries imported")
```

```python
# Cell 3: Load Multimodal Models
clip_model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")

sd_pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1",
    torch_dtype=torch.float16
)
sd_pipe = sd_pipe.to("cuda" if torch.cuda.is_available() else "cpu")

print("✓ CLIP and Stable Diffusion models loaded")
```

```python
# Cell 4: Text-to-Image Generation
prompts = [
    "A futuristic robot in a cyberpunk city",
    "An oil painting of a serene lake",
    "Abstract geometric art with vibrant colors",
]

fig, axes = plt.subplots(1, len(prompts), figsize=(15, 5))

for idx, prompt in enumerate(prompts):
    print(f"Generating {idx+1}/{len(prompts)}: {prompt}")
    image = sd_pipe(prompt).images[0]
    axes[idx].imshow(image)
    axes[idx].set_title(prompt, fontsize=10)
    axes[idx].axis('off')

plt.tight_layout()
plt.savefig("multimodal_results.png", dpi=300)
plt.show()

print("✓ Images generated!")
```

```python
# Cell 5: Image-to-Text (CLIP Analysis)
def analyze_image_with_text(image_path, text_queries):
    image = Image.open(image_path)
    
    inputs = clip_processor(
        text=text_queries,
        images=image,
        return_tensors="pt",
        padding=True
    )
    
    outputs = clip_model(**inputs)
    logits_per_image = outputs.logits_per_image
    probs = logits_per_image.softmax(dim=1)[0]
    
    results = {query: prob.item() for query, prob in zip(text_queries, probs)}
    return results

# Test (uncomment when you have an image)
# text_queries = ["a cat", "a dog", "a bird", "a car", "a person"]
# results = analyze_image_with_text("test.jpg", text_queries)
# for query, prob in sorted(results.items(), key=lambda x: x[1], reverse=True):
#     print(f"{query}: {prob:.2%}")

print("✓ Image analysis function ready")
```
