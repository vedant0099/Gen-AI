#### Setup and Imports
Imports necessary libraries (random, collections, nltk) and downloads the 'punkt' tokenizer.

````python
import random
import nltk
from collections import defaultdict
from nltk.tokenize import word_tokenize, sent_tokenize

nltk.download('punkt')
````

#### Build Markov Chain
Defines a function that takes input text, tokenizes it into sentences and words, and builds a Markov Chain dictionary mapping state tuples to possible next words.

````python
def build_markov_chain(text, order=2):
    tokens = []
    for sentence in sent_tokenize(text.lower()):
        tokens.extend(word_tokenize(sentence))
    
    markov_chain = defaultdict(list)
    
    for i in range(len(tokens) - order):
        state = tuple(tokens[i:i + order])
        next_word = tokens[i + order]
        markov_chain[state].append(next_word)
    
    return markov_chain
````

#### Load Data and Train Model
Reads content from `input.txt` and uses it to build the Markov Chain model with an order of 2.

````python
with open("./input.txt", "r", encoding="utf-8") as f:
    text = f.read()

model = build_markov_chain(text, order=2)
````

#### Generate Text Function
Defines a function to generate text by starting with a random state and choosing the next word based on probabilities derived from the Markov Chain.

````python
# Cell 5: Text generation
def generate_text(model, order=2, length=30):
    state = random.choice(list(model.keys()))
    generated = list(state)
    
    for _ in range(length):
        next_words = model.get(state)
        if not next_words:
            break
        next_word = random.choice(next_words)
        generated.append(next_word)
        state = tuple(generated[-order:])
    
    return " ".join(generated)
````

#### Output Samples
Generates and prints 3 samples of generated text to demonstrate the model's output.

````python
for i in range(3):
    print(f"Sample {i+1}:")
    print(generate_text(model, order=2, length=20))
    print()
````


It tries to **learn patterns from text** and then **generate new text** that looks similar.

This is done using a **Markov Chain**.

Meaning:

* Next word depends only on the **previous word(s)**
* No understanding, only probability

## 2. Tokenization

There is a function like:

```python
def tokenize(text):
    return text.split()
```

What it does:

* Takes raw text
* Splits it into words
* Example:

  ```
  "I love AI"
  → ["I", "love", "AI"]
  ```

Very basic. No punctuation handling. No cleaning.

---

## 3. Building the Markov Chain

Function:

```python
build_markov_chain(tokens, order=1)
```

What happens:

* `order=1` → looks at **1 word at a time**
* Creates a dictionary like:

```
{
  "I": ["love", "am"],
  "love": ["AI", "coding"]
}
```

Meaning:

* After `"I"` → `"love"` or `"am"` may come
* After `"love"` → `"AI"` or `"coding"`

This is the **core learning step**.

---

## 4. Training phase

You read a file like `input.txt` and do:

```python
tokens = tokenize(text)
chain = build_markov_chain(tokens)
```

This is not ML training.
No loss.
No gradient.
Just **counting word transitions**.

---

## 5. Text generation

There is a function like:

```python
generate_text(chain, start_word, length)
```

What it does:

1. Start from one word
2. Randomly pick the next word from the chain
3. Repeat until length is reached

Output:

* Grammatically weak
* Sometimes nonsense
* But statistically similar to input

---

## 6. Why it works (and why it’s weak)

### Works because:

* Language has local patterns
* Markov chains capture short-term structure

### Weak because:

* No memory beyond `order`
* No meaning
* Repeats loops
* Breaks on small datasets
