```python
# Cell 2: Import Libraries
from moviepy.editor import ImageSequenceClip, concatenate_videoclips
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
import numpy as np

print("✓ Libraries imported")
```

```python
# Cell 3: Setup Image Generator
pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

print("✓ Image generator ready")
```

```python
# Cell 4: Generate Frame Sequence
prompts = [
    "A sunrise over mountains, frame 1",
    "The sun rising higher over mountains, frame 2",
    "Bright daylight over mountains, frame 3",
    "Sun setting over mountains, frame 4",
    "Dusk over mountains, frame 5",
]

frames = []
for idx, prompt in enumerate(prompts):
    print(f"Generating frame {idx+1}/{len(prompts)}: {prompt}")
    image = pipe(prompt).images[0]
    image.save(f"frame_{idx:03d}.png")
    frames.append(np.array(image))
    print(f"  ✓ Saved frame_{idx:03d}.png")

print("\n✓ All frames generated!")
```

```python
# Cell 5: Create Video
clip = ImageSequenceClip(frames, fps=2)  # 2 frames per second
clip.write_videofile("animation.mp4", codec="libx264")

print("✓ Video saved as 'animation.mp4'")
```
